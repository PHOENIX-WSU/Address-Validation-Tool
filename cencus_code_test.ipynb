```python
import pandas as pd
```

```python
df = pd.read_csv('geocoded_results.csv', dtype = {'Unique ID': str, 'Input Address': str, 'Match Status': str, 'Match Type': str, 'Matched Address': str, 'Coordinates': str, 'TIGER Line ID': str, 'Side': str})
df.head()
```

```python
df['State FIPS'] = df['State FIPS'].fillna(0)
df['State FIPS'] = df['State FIPS'].astype(int)

df['County FIPS'] = df['County FIPS'].fillna(0)
df['County FIPS'] = df['County FIPS'].astype(int)

df['Tract'] = df['Tract'].fillna(0)
df['Tract'] = df['Tract'].astype(int)

df['Block'] = df['Block'].fillna(0)
df['Block'] = df['Block'].astype(int)

df['State FIPS'] = df['State FIPS'].astype(str)
df['County FIPS'] = df['County FIPS'].astype(str)
df['Tract'] = df['Tract'].astype(str)
df['Block'] = df['Block'].astype(str)
```

```python
df.head()
```

```python
df['fips_11'] = df.apply(lambda row: f"{row['State FIPS']}{row['County FIPS']}{row['Tract']}", axis=1)
df['fips_15'] = df.apply(lambda row: f"{row['State FIPS']}{row['County FIPS']}{row['Tract']}{row['Block']}", axis=1)
```
